T-Test between gender and Income

```python
df.head()
```

```python
male_Income=df.groupby('Gender').get_group('Male')['Income']
male_Income
```

```python
female_Income=df.groupby('Gender').get_group('Female')['Income']
female_Income
```

```python
sns.boxplot(data=df,x='Gender',y='Income');
```

```python
from scipy.stats import ttest_ind
```

```python
# assumption:
# Ho: There is no significant difference between male avg Income and female Avg Income
# Ha: There is signifiant difference between male avg income and female avg income
alpha=0.05
```

```python
t_stat,p_value=ttest_ind(male_Income,female_Income)
print(t_stat,p_value)
```

```python
if p_value< alpha:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```

```python
# We can say that there is significance diff between male avg income and female avg income.
```

```python
#assumption:
# Ho: There is no significant difference between male avg Income and female Avg Income
# Ha: male avg income  is greater than female avg income
alpha=0.05
```

```python
t_stat,p_value=ttest_ind(male_Income,female_Income,alternative='greater')
print(t_stat,p_value)
```

```python
if p_value< alpha:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```

```python
# Through this we can say male income is higher than female.
```

```python
#assumption:
# Ho: There is no significant difference between male avg Income and female Avg Income
# Ha: male avg income  is lesser than female avg income
alpha=0.05
```

```python
t_stat,p_value=ttest_ind(male_Income,female_Income,alternative='less')
print(t_stat,p_value)
```

```python
if p_value< alpha:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```
